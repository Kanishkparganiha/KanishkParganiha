```python
df['Date'] = pd.to_datetime(df.Date)



df['Date'].head()
```

```python
# Reset our index so datetime_utc becomes a column
df.reset_index(inplace=True)

# Create new columns

df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df.head()
```

```python
Month = {1: 'January', 2: 'Feb', 3: 'March',4: 'April',5: 'May',6: 'June',7: 'July',8: 'August',9: 'September',10: 'October',11: 'November',12: 'December'}
df['Month'] = df['Month'].map(Month)
```

```python
df.head()
```

# Data Cleaning

## 1. Dropping Location Coordinates column 

```python
df=df.drop(columns=['Date'])
df.head()
```

## 2. Date Column Filtering

```python
df['Date']=df.Date.apply(lambda x: x.replace('12:00:00 AM',''))
df.head()
```

```python
df=df.drop(columns=['Port Code'])
df.head()
```

# Statistical Description

## 1. Month-Wise Statistical Description

### 1. Ordinary Least Square Model

```python
R1 = ols('Value ~ C(Month)', data=df).fit()
R1.summary()
```

### 2. Anova Table

```python
Av1=sm.stats.anova_lm(R1, typ=1)
Av1
```

### 3. Post Tukey HUC Test

```python
vs = MultiComparison(df['Value'], df['Month'])
result = vs.tukeyhsd()
 
print(result)
print(vs.groupsunique)
```
