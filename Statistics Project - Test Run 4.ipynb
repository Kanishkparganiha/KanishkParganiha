# Inferential Statistics

## A. State-Wise Description

### 1. Ordinary Least Square Model

```python
R1 = ols('Value ~ C(State)', data=df).fit()
R1.summary()
```

### 2. Anova Table

```python
Av1=sm.stats.anova_lm(R1, typ=1)
Av1
```

### 3. Post Tukey HUC Test

```python
vs = MultiComparison(df['Value'], df['State'])
result = vs.tukeyhsd()
 
print(result)
print(vs.groupsunique)
```

## B. Transportation Wise Description

### 1. Ordinary Least  Square Model

```python
R2 = ols('Value ~ C(Measure)', data=df).fit()
R2.summary()
```

### 2. Anova Table

```python
Av2=sm.stats.anova_lm(R2, typ=1)
Av2
```

### 3. Post Tukey HUC Test

```python
vm = MultiComparison(df['Value'], df['Measure'])
result = vm.tukeyhsd()
 
print(result)
print(vm.groupsunique)
```

## C. Border Wise Description

### 1. Ordinary Least Square Model

```python
R3 = ols('Value ~ C(Border)', data=df).fit()
R3.summary()
```

### 2. Anova Table

```python
Av3=sm.stats.anova_lm(R3, typ=1)
Av3
```

### 3. Post Tukey HUC Test

```python
vb = MultiComparison(df['Value'], df['Border'])
result = vb.tukeyhsd()
 
print(result)
print(vb.groupsunique)
```
